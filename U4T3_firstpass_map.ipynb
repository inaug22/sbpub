**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
def plotinfo_byindex(index):
    x = prices.columns[2:]
    # iloc[row, col]
    y = prices.iloc[index,2:]
    name = str(prices['Borough_Name'].loc[prices.index[index]])
    return (x, y, name)

def plotinfo_byname(name):
    i = prices.Borough_Name[prices.Borough_Name == name].index.tolist()[0]
    return plotinfo_byindex(i)

def plotinfo_yearly(index):
    y = prices.iloc[index,2:]
    x = prices.columns[2:]
    # slice every 12 of y, store the average in Y
    Y =  [np.mean(n) for n in y[::12]]
    #  pick all mid-year points (June), store in X
    X = [n for n in x[5::12]]
    # account for more means than Junes (don't use less than half-year means)
    if(len(Y)>len(X)): Y.pop()
    # TODO: combine plotinfo functions to one with variable length args (optional arg: 'step' size)
    name = str(prices['Borough_Name'].loc[prices.index[index]])
    return (X, Y, name)

def addplot(plt, boroughs, interval):
    for i in boroughs:
        # picking the borough by index
        if(interval=='year'):
            data = plotinfo_yearly(i)
        # TODO: use elif, then else -> error handling
        else:
            data = plotinfo_byindex(i)
        # create plot
        plt.plot(data[0], data[1], linewidth=2, label=data[2]) # use s=2 for scatter, c='g' to set color

def adjust_axes(plt):        
    # add title and axes labels
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('House Price in £', fontsize=12)
    adjust_ticks(plt)

def adjust_ticks(plt):       
    x_tick_val = ['Jun 1995','Jun 2000', 'Jun 2005', 'Jun 2010', 'Jun 2015', 'Jun 2020']
    x_tick_lab = ['1995','2000', '2005', '2010', '2015', '2020']
    plt.xticks(x_tick_val, x_tick_lab)
    # TODO: figure out how to deal with auto adjusting units, e.g., 1e6 once data > 1 mil
    #y_tick_val = ['200000','400000', '600000', '800000']
    #y_tick_lab = ['200 k','400 k', '600 k', '800 k']
    #plt.yticks(y_tick_val, y_tick_lab)

def plot_boroughs(interval='monthly'):
    # general setup before creating plot
    # NOTE: rcParams don't seem to work with seaborn style
    # plt.rcParams['figure.figsize'] = [15, 10]
    plt.figure(figsize=(15,15), frameon=False, clear=True)
    plt.style.use('seaborn')

    # pick some rows
    boroughs = []
    for i in range(0, 32, 6):
        boroughs.append(i)

    # add plots 
    addplot(plt, boroughs, interval)

    # adjust axes
    adjust_axes(plt)
    
    # set title
    plt.title('House Price Development in Selected London Boroughs', fontsize=16, fontweight='bold')

    # add legend
    # NOTE: tried, but didn't work: labelcolor='linecolor'
    plt.legend(fontsize=12, handlelength=4) 

    # shop plot
    plt.show()

plot_boroughs()
```

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

```python
def calc_growth_rate(index, start_col, end_col):
    # TODO: ensure data goes that far
    start_mean = np.mean(prices.iloc[index, start_col:start_col+12])
    end_mean = np.mean(prices.iloc[index, end_col:end_col+12])
    return end_mean/start_mean

def plot_growth_rate(x, y, start, end):
    plt.figure(figsize=(10,15), frameon=False, clear=True)
    plt.stem(list(reversed(x)), list(reversed(y)), orientation='horizontal', markerfmt='None')
    # plt.gca().invert_yaxis()
    plt.xlabel("Growth Rate ({} - {}), using the method 'Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.')".format(start, end), fontsize=14)
    plt.ylabel('Borough', fontsize=14)
    plt.xticks(fontsize=11)
    plt.yticks(fontsize=11)
    plt.title('House Price Growth Rates in London Boroughs' , fontsize=16, fontweight='bold')
    plt.show()

def show_growth_rate(start_y, end_y, borough=''):
    start_col = prices.columns.get_loc('Jan ' + str(start_y))
    end_col = prices.columns.get_loc('Dec ' + str(end_y))
    # print string showing growth rate for given borough
    if(borough != ''):
        # get row number for borough - boroughs are unique, should be only one
        index = prices.index[prices.Borough_Name == borough].tolist()[0]
        print("Growth rate for " + borough + " is: " + "{:.2f}".format(calc_growth_rate(index-1, start_col, end_col)))
    # show stem plot comparing all boroughs
    else:
        x = []
        y = []
        for i in range(len(prices.Borough_Name)):
            x.append(prices.Borough_Name[i+1])
            y.append(calc_growth_rate(i, start_col, end_col))
        plot_growth_rate(x, y, start_y, end_y)
            
# show growth rate for a single borough - prints out string
show_growth_rate(1998, 2018, 'Harrow')
show_growth_rate(1998, 2018, 'Hackney')
# show growth rate for all boroughs between 1998 and 2018 - stem plot
show_growth_rate(1998, 2018)
```
